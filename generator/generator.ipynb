```python
import os

def dicoForm():
    dico = {}
    dico["text"] = '''<div class="field">
    <label for="[MinCol]" class="label">
        [MajCol]
    </label>
    <div class="control">
        <input type="text" class="input" id="[MinCol]" name="[MinCol]" [(ngModel)]="new[MajClass].[MinCol]" required />
        <span class="help is-color-danger-40">Help</span>
    </div>
</div>'''

    dico["number"] = '''<div class="field">
    <label for="[MinCol]" class="label">
        [MajCol]
    </label>
    <div class="control">
        <input type="number" class="input" id="[MinCol]" name="[MinCol]" [(ngModel)]="new[MajClass].[MinCol]" required />
        <span class="help is-color-danger-40">Help</span>
    </div>
</div>'''

    dico["date"] = '''<div class="field">
    <label for="[MinCol]" class="label">
        [MajCol]
    </label>
    <div class="control">
        <input type="date" class="input" id="[MinCol]" name="[MinCol]" [(ngModel)]="new[MajClass].[MinCol]" required />
        <span class="help is-color-danger-40">Help</span>
    </div>
</div>'''

    dico["datetime"] = '''<div class="field">
    <label for="[MinCol]" class="label">
        [MajCol]
    </label>
    <div class="control">
        <input type="datetime-local" class="input" id="[MinCol]" name="[MinCol]" [(ngModel)]="new[MajClass].[MinCol]" required />
        <span class="help is-color-danger-40">Help</span>
    </div>
</div>'''

    dico["time"] = '''<div class="field">
    <label for="[MinCol]" class="label">
        [MajCol]
    </label>
    <div class="control">
        <input type="time" class="input" id="[MinCol]" name="[MinCol]" [(ngModel)]="new[MajClass].[MinCol]" required />
        <span class="help is-color-danger-40">Help</span>
    </div>
</div>'''

    dico["select"] = '''<div class="field">
    <label for="[MinCol]" class="label">
        [MajCol]
    </label>
    <div class="control">
        <div class="select is-fullwidth">
            <select id="[MinCol]" name="[MinCol]" [(ngModel)]="new[MajClass].[MinCol]" >
                <option *ngIf="!new[MajClass].[MinCol]" value="" disabled selected>Selectionner [MajCol]</option>
                <option *ngFor="let row of list" [value]="row._id">
                    {{ row.nom }}
                </option>
            </select>
        </div>
        <span class="help is-color-danger-40">Help</span>
    </div>
</div>'''

    dico["radio"] = '''<div class="field">
    <label for="[MinCol]" class="label">
        [MajCol]
    </label>
    <div class="control" *ngFor="let row of list">
        <label class="radio">
            <input type="radio" name="[MinCol]" [value]="row._id" [(ngModel)]="new[MajClass].[MinCol]" [checked]="row._id === new[MajClass].[MinCol]"/>
            {{ row.nom }}
        </label>
        <span class="help is-color-danger-40">Help</span>
    </div>
</div>'''

    dico["checkbox"] = '''<div class="field">
    <label for="[MinCol]" class="label">
        [MajCol]
    </label>
    <div class="control" *ngFor="let row of list">
        <label for="checkbox" class="checkbox">
            <input type="checkbox" name="[MinCol][]" [value]="row._id" (change)="toggle[MajClass](row._id, $event)" [(ngModel)]="new[MajClass].[MinCol]" [checked]="new[MajClass].[MinCol]?.includes(row._id)"/>
            {{ row.nom }}
        </label>
        <span class="help is-color-danger-40">Help</span>
    </div>
</div>'''

    dico["textarea"] = '''<div class="field">
    <label for="[MinCol]" class="label">
        [MajCol]
    </label>
    <div class="control">
        <textarea class="textarea" id="[MinCol]" name="[MinCol]" [(ngModel)]="new[MajClass].[MinCol]" required ></textarea>
        <span class="help is-color-danger-40">Help</span>
    </div>
</div>'''
    return dico

def dicoList():
    dico = {}
    dico["th"] = '''<th class="has-text-left">[MajCol]</th>'''
    dico["td"] = '''<td>{{ [MinClass].[MinCol] }}</td>'''
    return dico

def read(filename):
    with open(filename, 'r') as fichier:
        contenu = fichier.read()
    return contenu

def generate_Service(chemin, classe):
    spec = read("services/class.service.spec.ts")
    service = read("services/class.service.ts")

    MajClass = classe.title()
    MinClass = classe.lower()

    service = service.replace('[MajClass]', MajClass)
    service = service.replace('[MinClass]', MinClass)

    spec = spec.replace('[MajClass]', MajClass)
    spec = spec.replace('[MinClass]', MinClass)

    spec_name = MinClass + '.service.spec.ts'
    service_name = MinClass + '.service.ts'

    chemin = chemin + 'services/'
    os.makedirs(chemin, exist_ok=True)

    with open( chemin + service_name, 'w') as fichier_modifie:
        fichier_modifie.write(service)
    with open( chemin + spec_name, 'w') as fichier_modifie:
        fichier_modifie.write(spec)

def generate_List(chemin, role, classe, colonne):
    chemin = chemin + role
    html = read("components/class/class-list.component.html")
    spec = read("components/class/class-list.component.spec.ts")
    ts = read("components/class/class-list.component.ts")
    dico = dicoList()

    MajClass = classe.title()
    MinClass = classe.lower()
    th = ""
    td = ""

    for col in colonne:
        column = col.lower()
        th = th + dico["th"] + "\n"
        th = th.replace("[MajCol]", column.title())
        td = td + dico["td"] + "\n"
        td = td.replace("[MinCol]", column)

    html = html.replace("[th]", th)
    html = html.replace("[td]", td)
    html = html.replace('[MajClass]', MajClass)
    html = html.replace('[MinClass]', MinClass)

    spec = spec.replace('[MajClass]', MajClass)
    spec = spec.replace('[MinClass]', MinClass)

    ts = ts.replace('[MajClass]', MajClass)
    ts = ts.replace('[MinClass]', MinClass)

    html_name = MinClass + '-list.component.html'
    spec_name = MinClass + '-list.component.spec.ts'
    ts_name = MinClass + '-list.component.ts'

    chemin = chemin + 'components/' + MinClass + '/'
    os.makedirs(chemin, exist_ok=True)

    with open( chemin + html_name, 'w') as fichier_modifie:
        fichier_modifie.write(html)
    with open( chemin + spec_name, 'w') as fichier_modifie:
        fichier_modifie.write(spec)
    with open( chemin + ts_name, 'w') as fichier_modifie:
        fichier_modifie.write(ts)

def generate_Create(chemin, role, classe, colonne, type):
    chemin = chemin + role
    html = read("components/class/class-create.component.html")
    spec = read("components/class/class-create.component.spec.ts")
    ts = read("components/class/class-create.component.ts")
    dico = dicoForm()

    MajClass = classe.title()
    MinClass = classe.lower()
    input = ""
    condition = ""
    data = ""

    for i, col in enumerate (colonne):
        if(i != 0):
            condition = condition + " && "
        column = col.lower()        
        input = input + dico[type[i]] + "\n"
        input = input.replace("[MajCol]", column.title())
        input = input.replace("[MinCol]", column)
        data = data + " " + column + ": '',"  
        condition = condition + "this.new" + MajClass + "." + column

    html = html.replace("[input]", input)
    html = html.replace('[MajClass]', MajClass)
    html = html.replace('[MinClass]', MinClass)

    spec = spec.replace('[MajClass]', MajClass)
    spec = spec.replace('[MinClass]', MinClass)

    ts = ts.replace('[MajClass]', MajClass)
    ts = ts.replace('[MinClass]', MinClass)
    ts = ts.replace('[data]', data)
    ts = ts.replace('[condition]', condition)

    html_name = MinClass + '-create.component.html'
    spec_name = MinClass + '-create.component.spec.ts'
    ts_name = MinClass + '-create.component.ts'

    chemin = chemin + 'components/' + MinClass + '/'
    os.makedirs(chemin, exist_ok=True)

    with open( chemin + html_name, 'w') as fichier_modifie:
        fichier_modifie.write(html)
    with open( chemin + spec_name, 'w') as fichier_modifie:
        fichier_modifie.write(spec)
    with open( chemin + ts_name, 'w') as fichier_modifie:
        fichier_modifie.write(ts)

def generate_Edit(chemin, role, classe, colonne, type):
    chemin = chemin + role
    html = read("components/class/class-edit.component.html")
    spec = read("components/class/class-edit.component.spec.ts")
    ts = read("components/class/class-edit.component.ts")
    dico = dicoForm()

    MajClass = classe.title()
    MinClass = classe.lower()
    input = ""
    condition = ""

    for i, col in enumerate (colonne):
        if(i != 0):
            condition = condition + " && "
        column = col.lower()   
        if(type[i] == "checkbox"):
            type[i] = "select"
        input = input + dico[type[i]] + "\n"
        input = input.replace("[MajCol]", column.title())
        input = input.replace("[MinCol]", column) 
        condition = condition + "this.new" + MajClass + "." + column

    html = html.replace("[input]", input)
    html = html.replace('[MajClass]', MajClass)
    html = html.replace('[MinClass]', MinClass)

    spec = spec.replace('[MajClass]', MajClass)
    spec = spec.replace('[MinClass]', MinClass)

    ts = ts.replace('[MajClass]', MajClass)
    ts = ts.replace('[MinClass]', MinClass)
    ts = ts.replace('[condition]', condition)

    html_name = MinClass + '-edit.component.html'
    spec_name = MinClass + '-edit.component.spec.ts'
    ts_name = MinClass + '-edit.component.ts'

    chemin = chemin + 'components/' + MinClass + '/'
    os.makedirs(chemin, exist_ok=True)

    with open( chemin + html_name, 'w') as fichier_modifie:
        fichier_modifie.write(html)
    with open( chemin + spec_name, 'w') as fichier_modifie:
        fichier_modifie.write(spec)
    with open( chemin + ts_name, 'w') as fichier_modifie:
        fichier_modifie.write(ts)

def addToRoute(chemin, role, classe):
    route = read(chemin + 'app.routes.ts')

    imports = '''import { [MajClass]ListComponent } from './components/[role][MinClass]/[MinClass]-list.component';
import { [MajClass]CreateComponent } from './components/[role][MinClass]/[MinClass]-create.component';
import { [MajClass]EditComponent } from './components/[role][MinClass]/[MinClass]-edit.component';
//new_import'''

    path = '''{ path: '[MinClass]s-edit/:id', component: [MajClass]EditComponent },
    { path: '[MinClass]s-create', component: [MajClass]CreateComponent }, 
    { path: '[MinClass]s', component: [MajClass]ListComponent },
    //new_path'''

    MajClass = classe.title()
    MinClass = classe.lower()

    route = route.replace('//new_import', imports)
    route = route.replace('//new_path', path)
    route = route.replace('[role]', role)
    route = route.replace('[MajClass]', MajClass)
    route = route.replace('[MinClass]', MinClass)

    with open( chemin + 'app.routes.ts', 'w') as fichier_modifie:
        fichier_modifie.write(route)
```

```python
chemin = 'D:/ITU/S7/web_avance/mean_project/frontend/src/app/'
# chemin = 'D:/ITU/S7/web_avance/mean_project/generator/exemple/'
classe = 'poste'
colonne = [ "idemploye", "idrole"]
type = [ "select", "checkbox"]
role = ''

generate_Service(chemin, classe)
generate_List(chemin, role, classe, colonne)
generate_Create(chemin, role, classe, colonne, type)
generate_Edit(chemin, role, classe, colonne, type)
addToRoute(chemin, role, classe)
```
